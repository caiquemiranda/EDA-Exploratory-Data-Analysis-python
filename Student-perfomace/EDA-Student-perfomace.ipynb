```python
# importação das bibliotecas

import pandas as pd                                                     # importação da biblioteca pandas
import numpy as np                                                      # importação da biblioteca numpy
import matplotlib                                                       # importação da biblioteca matplotlib
import seaborn as sns                                                   # importação da biblioteca seaborn
import matplotlib.pyplot as plt                                         # importação da biblioteca matplotlib.pyplot
from pylab import mpl, plt                                              # importação da biblioteca pylab
```

```python
# leitura de dados direto do github

students = pd.read_csv('https://raw.githubusercontent.com/caiquemiranda/EDA-Exploratory-Data-Analysis/main/data/StudentsPerformance.csv')
```

```python
# configurações do estilo do gráfico

plt.rcParams['figure.figsize'] = (16,10)
plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'
```

```python
# grafição de distribuição dos dados em grafico de barras

plt.subplot(221)
students['gender'].value_counts().plot(kind='bar', title='Gender of students', figsize=(10, 7))
plt.xticks(rotation=0)

plt.subplot(222)
students['race/ethnicity'].value_counts().plot(kind='bar', title='Race/ethnicity of students')
plt.xticks(rotation=0)

plt.subplot(223)
students['lunch'].value_counts().plot(kind='bar', title='Lunch status of students')
plt.xticks(rotation=0)

plt.subplot(224)
students['test preparation course'].value_counts().plot(kind='bar', title='Test preparation course')
plt.xticks(rotation=0);
```

```python
# visualização de distribuição dos dados em grafico de pizza 

plt.subplot(221)
students['gender'].value_counts().plot(kind='pie', title='Gender of students', autopct='%1.1f%%', figsize=(10, 7))
plt.xticks(rotation=0)

plt.subplot(222)
students['race/ethnicity'].value_counts().plot(kind='pie', title='Race/ethnicity of students', autopct='%1.1f%%')
plt.xticks(rotation=0)

plt.subplot(223)
students['lunch'].value_counts().plot(kind='pie', title='Lunch status of students', autopct='%1.1f%%')
plt.xticks(rotation=0)

plt.subplot(224)
students['test preparation course'].value_counts().plot(kind='pie', title='Test preparation course', autopct='%1.1f%%')
plt.xticks(rotation=0);

plt.show()
```

```python
# visualização de variveis numericas em boxplot

students.boxplot()
plt.show()
```

```python
# grafico de distribuição com densidade do score 'math score'

sns.displot(students['math score'], kde=True, bins = 20)
plt.show()
```

```python
# grafico de distribuição com densidade do score 'reading score'

sns.displot(students['reading score'], kde=True)
plt.show()
```

```python
# grafico de distribuição com densidade do score 'writing score'

sns.displot(students['writing score'], kde=True)
plt.show()
```

```python
# correlação entre as variaveis score 'math score', 'reading score' e 'writing score'

corr =students.corr()
sns.heatmap(corr, annot=True,square=True)
plt.yticks(rotation=0)
plt.show()
```

```python
# grafiicos de dispersão separando por cores atribuidas ao genero com relação entre as variaveis 'math score' e 'writing score'

sns.relplot(x='math score', y='writing score', hue = 'gender', data=students)
plt.show()
```

```python
# visualização grafica do nivel de escolaridade dos pais

students.groupby('parental level of education')[['math score',  'reading score', 'writing score']].mean().T.plot(figsize = (12, 6))
plt.show()
```

```python
# visualização grafica da performance nos diferentes tipos de testes

students.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean().T.plot(kind='barh', figsize=(12,6))
plt.show()
```
